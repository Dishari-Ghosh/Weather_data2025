```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kolkata_data.csv')
```

```python
df.head()
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
df = df.dropna()
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
df['Month'] = df['DATE'].dt.month
df['Year'] = df['DATE'].dt.year
```

```python
plt.figure(figsize=(15,6))
plt.plot(df['DATE'], df['TAVG'], label='Temperature (C)')
plt.title('Daily average Temperature Trends in Kolkata')
plt.xlabel('Date')
plt.ylabel('Average_Temperature (C)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['PRCP'], kde=True)
plt.title('Rainfall Distribution in Kolkata')
plt.xlabel('Rainfall (mm)')
plt.xlim(0,0.2) 
plt.show()
```

```python
corr_matrix = df[['PRCP', 'TAVG', 'TMAX', 'TMIN']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Weather Features')
plt.show()
```
